```python
#from cookies import cookies

import pandas as pd
from facebook_scraper import get_posts
import matplotlib.pyplot as plt
```

```python
# Initialize dataframe to scrape Facebook post
post_df_full = pd.DataFrame(columns = [])
```

```python
import cookies 

# Start to collect Facebook post data by facebook_scraper library
for post in get_posts('Samsung', cookie=None, extra_info=True, pages=5, options={"comments": True}):
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()
    post_df_full = post_df_full.append(fb_post_df)
    print(post['post_id']+' get')
```

```python
# Check dataframe information, 47 data columns of Facebook post are being scraped
post_df_full.info()
```

```python
# Display header of dataframe
post_df_full.head()
```

```python
# Save dataframe into csv file
post_df_full.to_csv(r'fb_scrapped_data_Samsung.csv', index = False)
```

```python
# Perform basic line plot to visualize post datetime vs likes
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(post_df_full['time'], post_df_full['likes'], marker="o")
from matplotlib.dates import DateFormatter
ax.xaxis.set_major_formatter(DateFormatter('%d-%m-%y %H:%M'))
```

```python
# Perform basic line plot to visualize post datetime vs likes, shares and comments
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(post_df_full['time'], post_df_full['likes'], label = "Likes", marker="o")
ax.plot(post_df_full['time'], post_df_full['shares'], label = "Shares", marker="s")
ax.plot(post_df_full['time'], post_df_full['comments'], label = "Comments", marker="*")
plt.legend()
from matplotlib.dates import DateFormatter
ax.xaxis.set_major_formatter(DateFormatter('%d-%m-%y %H:%M'))
```

```python
import re
#clean the text column
def clean_text(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text)  # Removing @mentions
    text = re.sub(r'#', '', text)  # Removing '#' hash tag
    text = re.sub(r'RT[\s]+', '', text)  # Removing RT
    text = re.sub(r'https?:\/\/\S+', '', text)  # Removing hyperlink
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # Removing punctuation
    text = re.sub(r'\s+', ' ', text)  # Removing extra whitespace
    text = text.lower()  # Converting to lowercase
    return text
```

# Sentiment Analysis

```python
import numpy as np
import pandas as pd
df1 = pd.read_csv('fb_scrapped_data_Samsung.csv' , encoding='utf-8')
df2 = pd.read_csv('fb_scrapped_data_Nintendo.csv' , encoding='utf-8')    
df3 = pd.read_csv('fb_scrapped_data_Sony.csv' , encoding='utf-8')
df4= pd.read_csv('insta_scrapped_data.xlsx' , encoding='utf-8')
df5 = pd.read_csv('insta_scrapped_data_2.xlsx' , encoding='utf-8')


df = pd.concat([df1, df2, df3 ,df4 ,df5 ], ignore_index=True)
df.head()
```

```python
import textblob
# classifying tweets as positive, negative and neutral
def get_sentiment(text):
    sentiment = textblob.TextBlob(text).sentiment.polarity
    if sentiment < 0:
        return 'Negative'
    elif sentiment == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
df_final['Sentiment'] = df_final['text'].apply(get_sentiment)
```

```python
#store positive in one csv file and negative in another csv file 
df_positive = df_final[df_final['Sentiment']=='Positive']
df_negative = df_final[df_final['Sentiment']=='Negative']
df_neutral = df_final[df_final['Sentiment']=='Neutral']

df_positive.to_csv('positive.csv', index=False)
df_negative.to_csv('negative.csv', index=False)
df_neutral.to_csv('neutral.csv', index=False)
```
